```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

# Apply modern professional theme
plt.style.use("seaborn-v0_8-darkgrid")

plt.figure(figsize=(16,12))

# ---------------- 1. Visitors vs Profit (Scatter) ----------------
plt.subplot(2,2,1)
plt.scatter(df["Visitors"], df["Profit"], s=120)
plt.title("Visitors vs Profit Relationship", fontsize=12, fontweight="bold")
plt.xlabel("Visitors")
plt.ylabel("Profit")

# ---------------- 2. Visitors Distribution (Box Plot) ----------------
plt.subplot(2,2,2)
plt.boxplot(df["Visitors"], patch_artist=True)
plt.title("Visitors Distribution", fontsize=12, fontweight="bold")
plt.ylabel("Visitors")

# ---------------- 3. Visitors Trend (Line) ----------------
plt.subplot(2,2,3)
plt.plot(df["Month"], df["Visitors"], marker="o", linewidth=3)
plt.title("Monthly Visitors Trend", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Visitors")

# ---------------- 4. Month-wise Visitors (Bar) ----------------
plt.subplot(2,2,4)
plt.bar(df["Month"], df["Visitors"])
plt.title("Month-wise Visitors", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Visitors")

# ---------------- Dashboard Title ----------------
plt.suptitle("Website Visitors Performance Dashboard", fontsize=18, fontweight="bold")

# Layout optimization
plt.tight_layout(rect=[0,0,1,0.95])
plt.savefig("companyA_report.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("company_sales.csv")

# Apply professional style
plt.style.use("seaborn-v0_8-darkgrid")

# Create big dashboard
plt.figure(figsize=(16,12))

# -------------------- 1. Company A Sales --------------------
plt.subplot(2,2,1)
plt.plot(df["Month"], df["CompanyA"], marker="o", linewidth=3)
plt.title("Company A Monthly Sales", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Sales")

# -------------------- 2. Company B Sales --------------------
plt.subplot(2,2,2)
plt.plot(df["Month"], df["CompanyB"], marker="s", linewidth=3)
plt.title("Company B Monthly Sales", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Sales")

# -------------------- 3. Visitors --------------------
plt.subplot(2,2,3)
plt.bar(df["Month"], df["Visitors"])
plt.title("Monthly Visitors", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Visitors")

# -------------------- 4. Profit --------------------
plt.subplot(2,2,4)
plt.plot(df["Month"], df["Profit"], marker="d", linewidth=3)
plt.title("Monthly Profit Trend", fontsize=12, fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Profit")

# -------------------- Dashboard Title --------------------
plt.suptitle(
    " Mobile Company Business Performance Dashboard",
    fontsize=18,
    fontweight="bold"
)

# Perfect spacing
plt.tight_layout(rect=[0,0,1,0.95])
plt.savefig("business_dashboard.png", dpi=300, bbox_inches="tight")

plt.show()
```
