```python
"""Ek chhoti si kahani socho

Tum ek shop chala rahi ho.

Tum roz note karti ho:

Kitne log aaye (Visitors)

Kitna sale hua (Sales)

Tumhare paas yeh data hai:

Visitors	Sales
100	2000
200	3000
300	5000
400	6000

Ab tum yeh jaan‚Äôna chahti ho:

‚ÄúZyada log aaye to kya sale zyada hota hai?‚Äù

Iska jawab:
‚ùå Bar graph nahi
‚ùå Line graph nahi
‚úÖ Scatter plot

üü¢ Scatter plot me kya hota hai?

Tum graph paper socho:

X-axis ‚Üí Visitors
Y-axis ‚Üí Sales

Ab har row ek dot ban jata hai:

(100, 2000)
(200, 3000)
(300, 5000)

Matlab har din = ek point

üü° Graph ka secret rule

Dots agar:
‚Üó upar ja rahe ‚Üí relation hai (visitors badhe = sales badhi)
‚Üò neeche ja rahe ‚Üí negative relation
üéØ idhar-udhar ‚Üí koi relation nahi

Bas yahi scatter plot ka kaam hai."""-8i
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

plt.scatter(df["Visitors"], df["CompanyA"])

plt.xlabel("Visitors")
plt.ylabel("Company A Sales")
plt.title("Visitors vs Company A Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Visitors"], df["CompanyA"], s=100, alpha=0.7)

plt.xlabel("Visitors")
plt.ylabel("Company A Sales")
plt.title("Relationship Between Visitors and Company A Sales")
plt.grid(True)

plt.show()
```
