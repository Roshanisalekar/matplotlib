```python
 """ Created a scatter plot using CSV data to analyze the relationship between website visitors and sales of two mobile 
 companies and identify which company converts visitors into sales more effectively."""
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

plt.figure(figsize=(8,5))

# Company A
plt.scatter(df["Visitors"], df["CompanyA"], s=100, label="Company A")

# Company B
plt.scatter(df["Visitors"], df["CompanyB"], s=100, label="Company B")

plt.xlabel("Visitors")
plt.ylabel("Sales")
plt.title("Visitors vs Sales Comparison of Two Companies")
plt.legend()
plt.grid(True)

plt.show()
```
