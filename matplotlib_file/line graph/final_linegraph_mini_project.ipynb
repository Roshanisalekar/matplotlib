```python
"""Mini Project: Mobile Company Sales Analysis (Python & Matplotlib)

Loaded real sales data from CSV using Pandas

Created line graphs to analyze monthly sales trends of two companies

Compared performance using legends, markers and grid

Identified growth patterns and best performing months"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")
df
```

```python

plt.plot(df["Month"],df["CompanyA"], color="blue", linestyle="--", marker="o", linewidth=2, label="Company A")
plt.plot(df["Month"], df["CompanyB"], color="green", linestyle="-", marker="s", linewidth=2, label="Company B")

plt.xlabel("Months")
plt.ylabel("Recharge Sales")
plt.title("Mobile Company Sales Comparison")
plt.legend()
plt.grid(True)
plt.show()

```
