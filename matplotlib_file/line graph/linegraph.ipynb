```python
'''6️⃣ Multiple Lines (Company Comparison)

Socho:
Company A vs Company B'''
```

```python

plt.figure(figsize=(10,6))
months=["jan","feb","mar"]
A=[2000,3000,2500]
B=[2000,1000,3000]
plt.plot(months,A,label="Company A")
plt.plot(months,B,label="Company B")
plt.legend()
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales in Rs")

plt.show()
```

```python

plt.figure(figsize=(10,6))
months=["jan","feb","mar"]
A=[2000,3000,2500]
B=[2000,1000,3000]
plt.plot(months,A,label="Company A",color="blue",linestyle=":",marker="o")
plt.plot(months,B,label="Company B",color="green",linestyle=":",marker="s")

plt.legend()
plt.title("Sales Comparison")
plt.xlabel("Months")
plt.ylabel("Sales in Rs")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar"]

A = [2000,3000,2500]
B = [1800,2800,2600]

plt.plot(months, A, color="blue", linestyle="--", marker="o", linewidth=2, label="Company A")
plt.plot(months, B, color="green", linestyle="-", marker="s", linewidth=2, label="Company B")

plt.xlabel("Months")
plt.ylabel("Sales (Rs)")
plt.title("Company Sales Comparison")
plt.legend()
plt.grid(True)

plt.show()
```

```python
'''
🧠 Deep Meaning of New Things
🟦 color

Line ka rang

color="blue"


Isse graph easy to read hota hai.

➖ linestyle

Line ka style:

"--"  → dashed  
"-"   → solid  
":"   → dotted  


Different lines easily pehchani jaati hain.

🔵 marker

Har point pe symbol

"o" → circle  
"s" → square  


Batata hai har month ka exact data point.

📏 linewidth

Line ki motai (thickness)

linewidth=2


Presentation ke liye important.

🧩 plt.grid(True)

Graph ke peeche grid aati hai
Isse reading easy hoti hai.

⭐ Legend (Most Important)
plt.legend()


Legend ka matlab hota hai:

"Kaunsi line kiske liye hai"

Graph ke corner me aisa box aayega:

🟦 Company A
🟥 Company B


Agar legend nahi likha:
➡ user ko pata hi nahi chalega kaunsi line kis company ki hai 😵

Isliye multiple lines ke saath legend compulsory hota hai

6️⃣ Title
plt.title("Sales Comparison")


Graph ka naam
Batata hai graph kis cheez ka hai

7️⃣ Show
plt.show()


Graph ko screen par display karta hai.'''
```
