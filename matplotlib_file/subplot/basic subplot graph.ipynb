```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

plt.subplot(1,2,1)
plt.plot(df["Month"], df["CompanyA"])
plt.title("Company A Sales")

plt.subplot(1,2,2)
plt.plot(df["Month"], df["CompanyB"])
plt.title("Company B Sales")

plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(df["Month"], df["CompanyA"])
plt.title("Company A")

plt.subplot(2,2,2)
plt.plot(df["Month"], df["CompanyB"])
plt.title("Company B")

plt.subplot(2,2,3)
plt.bar(df["Month"], df["Visitors"])
plt.title("Visitors")

plt.subplot(2,2,4)
plt.plot(df["Month"], df["Profit"])
plt.title("Profit")

plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn-v0_8")
plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
plt.plot(df["Month"], df["CompanyA"], marker="o", linewidth=2)
plt.title("Company A Sales")

plt.subplot(2,2,2)
plt.plot(df["Month"], df["CompanyB"], marker="s", linewidth=2)
plt.title("Company B Sales")
 
plt.subplot(2,2,3)
plt.bar(df["Month"], df["Visitors"])
plt.title("Monthly Visitors")

plt.subplot(2,2,4)
plt.plot(df["Month"], df["Profit"], marker="d", linewidth=2)
plt.title("Profit Trend")

plt.suptitle("Mobile Company Business Dashboard", fontsize=16, fontweight="bold")
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
