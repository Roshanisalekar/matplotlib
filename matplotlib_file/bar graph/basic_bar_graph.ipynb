```python
"""Bar Graph kab use hota hai?

Line graph trend dikhata hai.
Bar graph comparison dikhata hai.

üëâ Question:

‚ÄúWhich month had highest sales?‚Äù
‚ÄúWhich company sold more in May?‚Äù"""
```

```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")


plt.bar(df["Month"], df["CompanyA"])
plt.title("Company A Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.bar(df["Month"], df["CompanyB"])
plt.title("Company B Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
import numpy as np

months = df["Month"]
A = df["CompanyA"]
B = df["CompanyB"]

x = np.arange(len(months))

plt.bar(x-0.2, A, width=0.4, label="Company A")
plt.bar(x+0.2, B, width=0.4, label="Company B")

plt.xticks(x, months)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison (Bar Graph)")
plt.legend()
plt.show()

# xticks ka matlab:
# Numeric positions ko meaningful names dena
```

```python
"""XTICKS ka simple meaning:

X-axis par jo numbers hote hain unko hum apni marzi ke naam (labels) dete hain.

Easy words me:

xticks = X-axis ke labels set karna

Example:

By default graph me X-axis hota hai:

0   1   2   3   4


But tum chahti ho:

Jan  Feb  Mar  Apr  May


To hum likhte hain:

plt.xticks([0,1,2,3,4], ["Jan","Feb","Mar","Apr","May"])

Tumhare project me:
plt.xticks(x, months)


Matlab:

Jahan 0 hai wahan Jan likho
Jahan 1 hai wahan Feb likho
Jahan 2 hai wahan Mar likho
‚Ä¶

One-line Interview Answer:

xticks is used to set labels on the X-axis."""
```
