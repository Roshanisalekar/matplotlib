```python
"""Used Python and Matplotlib to create bar graphs from CSV data to compare monthly sales of two mobile companies and find the
best performing company each month."""
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("company_sales.csv")
```

```python
df
```

```python
Months=df["Month"]
A=df["CompanyA"]
B=df["CompanyB"]
x=np.arange(len(Months))
plt.bar(x-0.2,A,label="CompanyA",width=0.4)
plt.bar(x+0.2,B,label="CompanyB",width=0.4)
plt.xticks(x,Months)
plt.legend()
plt.grid(True)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison (Bar Graph)")
plt.show()
```
