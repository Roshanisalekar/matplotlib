```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

# Apply modern professional theme
plt.style.use("seaborn-v0_8-darkgrid")

plt.figure(figsize=(16,12))

plt.subplot(2,2,1)
months=df["Month"]
A=df["CompanyA"]
B=df["CompanyB"]
x=np.arange(len(months))
plt.bar(x-0.2,A,label="COMPANYA",width=0.4)
plt.bar(x+0.2,B,label="COMPANYB",width=0.4)
plt.xticks(x,months)
plt.grid()
plt.legend()


plt.subplot(2,2,2)
plt.pie(df["Profit"],labels=df["Month"],autopct="%1.1f%%",startangle=90,explode=[0.1,0,0,0,0,0])

plt.subplot(2,2,3)
plt.scatter(df["CompanyA"],df["Visitors"],s=100,label="companya")
plt.scatter(df["CompanyB"],df["Visitors"],s=100,label="companyb")
plt.legend()
plt.grid()


plt.subplot(2,2,4)
plt.plot(df["Month"],df["CompanyA"],marker="s",linewidth=3,linestyle="--",label="companya")
plt.plot(df["Month"],df["CompanyB"],marker="s",linewidth=3,linestyle="--",label="companyb")
plt.xlabel("months")
plt.ylabel("companys")
plt.title("company by sales")
plt.annotate(
    "Highest Sale",
    xy=("May", 5000),
    xytext=("May", 5500),
    arrowprops=dict(arrowstyle="->")
)

plt.legend()

plt.suptitle("Company Sales Dashbord",fontsize=20,fontweight="bold")
plt.tight_layout(rect=[0,0,1,0.95])
plt.savefig("com_sa_dashbord.png",dpi=300,bbox_inches="tight")
plt.show()
```
