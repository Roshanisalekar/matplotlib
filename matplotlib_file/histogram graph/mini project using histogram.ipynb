```python
# Built a histogram using CSV data to analyze monthly profit distribution of a mobile company and identify the most frequent profit ranges for business insights.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("company_sales.csv")
```

```python
df
```

```python
plt.hist(df["Profit"],bins=5,edgecolor="black")

plt.xlabel("Profit Range")
plt.ylabel("Number of Months")
plt.title("Monthly Profit Distribution of Mobile Company")

plt.grid(True)

plt.show()
```
