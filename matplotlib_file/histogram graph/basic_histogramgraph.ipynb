```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales.csv")

plt.figure(figsize=(8,5))

plt.hist(df["Profit"], bins=5, edgecolor="black")

plt.xlabel("Profit Range")
plt.ylabel("Number of Months")
plt.title("Monthly Profit Distribution of Mobile Company")

plt.grid(True)

plt.show()
```
