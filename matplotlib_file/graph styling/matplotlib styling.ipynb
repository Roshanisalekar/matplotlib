```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

df = pd.read_csv("company_sales.csv")

plt.boxplot(df["Profit"])
plt.title("Company Sales", fontsize=16, fontweight="bold")
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales", fontsize=12)


plt.grid(True)
plt.show()
```

```python
"""üîπ 4Ô∏è‚É£ Colors (Business standard)

Avoid bright colors ‚ùå
Use corporate colors:

plt.plot(df["Month"], df["CompanyA"], color="#1f77b4", label="Company A")
plt.plot(df["Month"], df["CompanyB"], color="#ff7f0e", label="Company B")


These are used by PowerBI & Tableau"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

df = pd.read_csv("company_sales.csv")

plt.plot(df["Month"], df["CompanyA"], color="#1f77b4", label="Company A")
plt.plot(df["Month"], df["CompanyB"], color="#ff7f0e", label="Company B")
plt.title("Company Sales", fontsize=16, fontweight="bold")
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")

df = pd.read_csv("company_sales.csv")

plt.figure(figsize=(12,6))

plt.plot(df["Month"], df["CompanyA"], marker="o", linewidth=2, label="Company A")
plt.plot(df["Month"], df["CompanyB"], marker="s", linewidth=2, label="Company B")

plt.title("Monthly Sales Trend", fontsize=16, fontweight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales", fontsize=12)

plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
